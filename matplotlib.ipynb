#Basic Plotting
The most important function in `matplotlib` is `plot`. `plo`t allows you to plot 2-dimensional data. 

Generally speaking, plotting in 2-dimensional space involves the following steps:

1. Create a sequence of $x$ values.
2. Create a sequence of $y$ values.
3. Use  ``plt.plot(x,y,[fmt], **kwargs)``

  *   ``[fmt]``: optinal format string
  *   ``**kwargs``: optinal keyword arguments specifying the properties of the plot
4. Add other features to the plot such as lables, title, legends, etc.
5. Use **``plt.show()``** to display the resulting plot.

##A very simple example



```python
# Run thic cell to see the result
x = [-5, -1, 0, 2, 3]
y = [2, -1, 1, -4, 3]
plt.plot(x, y)
plt.show()
```

## Another example

Here is another example: drawing a sine function


> using `arange` function of `np` (`numpy`) to generate an array of number starting from $0$, $0.1$, $0.2$ ... all the way to $< 3 * \pi$

> using ``np.sin()`` function to calculate sine values of the above values

```python
# run this cell to see the result
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

# Plot the points using matplotlib
plt.plot(x, y)
plt.show()
```

##Add title, legend, and axis labels

Here we draw two functions in one figure. 

```python
# Run this cell to see the result
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plotting
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
```

#Subplots 

We use **``plt.subplot``** function to plot multiple smaller figures in one bigger one. **``plt.subplot``** takes (at least) 3 inputs: $m$,  $n$ and $i$. It creates a figure with $m\times n$ grid of subplots and sets the $i\text{th}$ sublot as the current plot.

Example:

> ``plt.subplot(2, 1, 1)``: creates $2\times1$, i.e., two small subplots on top of each other, grid, and sets the top one as the current plot. 


> ``plt.subplot(2, 1, 2)``: sets the lower subplot as the current plot. 

Below is an example: let's plot sine and cosime functions on top of each other

```python
# Run this cell to see the result
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Create a subplot grid that has height 2 and width 1,
# and set the first subplot as the current plot.
plt.subplot(2, 1, 1)

# Plot the sine function
plt.plot(x, y_sin)
plt.title('Sine')
plt.subplots_adjust(hspace = 0.4) # Adjust space to display Title properly

# Set the second subplot as the current plot and plot the cosine function.
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosine')


# Show the figure.
plt.show()
```
